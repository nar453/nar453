Tabla interactiva con **Plotly**

```python
import matplotlib.pyplot as plt
import ast
from IPython.display import display
import numpy as np

# Paso 1: Filtrar por status
status_abierto = ['RELEASING', 'NOT_YET_RELEASED']
df_isin = df[df['status'].isin(status_abierto)].copy()

# Paso 2: Convertir columna a lista si es tipo string
# Suponiendo que la columna se llama 'genres'
# Use .str.split(',') to split the string into a list, handling potential NaN values
df_isin['genres'] = df_isin['genres'].apply(lambda x: x.split(',') if isinstance(x, str) else x)


# Paso 3: Explode para que cada género tenga su propia fila
df_exploded = df_isin.explode('genres')

# Paso 4: Mostrar tabla estilizada
def resaltar_status(val):
    if val in status_abierto:
        return 'background-color: lightblue'
    return ''

import matplotlib.pyplot as plt

# Crear gráfico de barras con ajustes
plt.figure(figsize=(12, 6))  # Aumenta tamaño horizontal de la gráfica

df_exploded['genres'].value_counts().plot(kind='bar', edgecolor='black')

plt.title('Animes con status abierto por género')
plt.xlabel('Género')
plt.ylabel('Número de animes')

plt.xticks(rotation=45, ha='right')  # Gira etiquetas y las alinea hacia la derecha
plt.tight_layout()  # Ajusta automáticamente para que no se encime nada

plt.show()
```

```python
import plotly.graph_objects as go


# Seleccionar columnas deseadas
# Use title_romaji or title_english if title_romaji is null
df_tabla = df_isin[['title_romaji', 'title_english', 'season', 'year', 'genres']].copy()
df_tabla['title'] = df_tabla['title_romaji'].fillna(df_tabla['title_english'])


# Crear tabla Plotly
fig = go.Figure(data=[go.Table(
    header=dict(
        values=["Título", "Temporada", "Año", "Géneros"],
        fill_color='lightblue',
        align='left',
        font=dict(size=12, color='black')
    ),
    cells=dict(
        values=[
            df_tabla['title'],
            df_tabla['season'],
            df_tabla['year'],
            df_tabla['genres']
        ],
        fill_color='white',
        align='left',
        height=25
    ))
])

fig.update_layout(
    width=1000,
    height=600,
    title="Tabla de animes con status abierto"
)

fig.show()
```
